```python
import zipfile
import re

def unzip_dataset(path=None, mode="r", output=None):
    pattern = r"(?:./)?(.+)\.\w+"
    if output == None and path is not None:
        output = f"./data/{re.search(pattern, path)[1]}"
    zipfile.ZipFile(path, mode).extractall(output)
    print(f"Stored at {output}")

train_zip = "./horse-or-human.zip"
val_zip = "./val-horse-or-human.zip"

unzip_dataset(path=train_zip)
unzip_dataset(path=val_zip)
```

```python
import os

train_horse_dir = os.path.join("./data/horse-or-human/horses")
train_human_dir = os.path.join("./data/horse-or-human/humans")
val_horse_dir = os.path.join("./data/val-horse-or-human/horses")
val_human_dir = os.path.join("./data/val-horse-or-human/humans")

train_horse_name = os.listdir(train_horse_dir)
train_human_name = os.listdir(train_human_dir)
val_horse_name = os.listdir(val_horse_dir)
val_human_name = os.listdir(val_human_dir)

total_train_horse_images = len(train_horse_name)
total_train_human_images = len(train_human_name)
total_train_images = total_train_horse_images + total_train_human_images

total_val_horse_images = len(val_horse_name)
total_val_human_images = len(val_human_name)
total_val_images = total_val_horse_images + total_val_human_images

show_name = 3

print(f"""
Total train horse: {total_train_horse_images}
Total train human: {total_train_human_images}
Total train images: {total_train_images}

Total val horse: {total_val_horse_images}
total val human: {total_val_human_images}
total val images: {total_val_images}

first {show_name} horse images name: {train_horse_name[:show_name]}
first {show_name} human images name: {train_human_name[:show_name]}
first {show_name} horse val images name: {val_horse_name[:show_name]}
first {show_name} human val images name: {val_human_name[:show_name]}
""")
```

## Data Preprocessing

### 150x150 pixels

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(rescale=1/255)
val_datagen = ImageDataGenerator(rescale=1/255)

train_gen_150px = train_datagen.flow_from_directory(
    "./data/horse-or-human/",
    target_size=(150,150),
    batch_size=128,
    class_mode="binary"
)

val_gen_150px = train_datagen.flow_from_directory(
    "./data/val-horse-or-human/",
    target_size=(150,150),
    batch_size=32,
    class_mode="binary"
)
```

```python
def get_img_path(dir_name, fname):
    """
    dir_name: path to the directory that contains images 
    fname: file_name
    """
    return [os.path.join(dir_name, x) for x in fname]

train_horse_img_arr = get_img_path(dir_name=train_horse_dir, fname=train_horse_name)
train_human_img_arr = get_img_path(dir_name=train_human_dir, fname=train_human_name)
val_horse_img_arr = get_img_path(dir_name=val_horse_dir, fname=val_horse_name)
val_human_img_arr = get_img_path(dir_name=val_human_dir, fname=val_human_name)

len(train_horse_img_arr), len(train_human_img_arr), len(val_horse_img_arr), len(val_human_img_arr)
```

## Build Model

### Model 1 - 3 convolution layers with 150x150px input shape

```python
import tensorflow as tf
```

```python
model_1 = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(16, (3,3), activation=tf.nn.relu, input_shape=(150,150,3)),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Conv2D(32, (3,3), activation=tf.nn.relu),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Conv2D(64, (3,3), activation=tf.nn.relu),
    tf.keras.layers.MaxPooling2D(2, 2),
    
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(512, activation=tf.nn.relu),
    tf.keras.layers.Dense(1, activation=tf.nn.sigmoid)
])

model_1.summary()
```

### Model 2 - 5 convolution layers with 150x150px input shape

```python
model_2 = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(16, (3,3), activation=tf.nn.relu, input_shape=(150,150,3)),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Conv2D(32, (3,3), activation=tf.nn.relu),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Conv2D(64, (3,3), activation=tf.nn.relu),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Conv2D(64, (3,3), activation=tf.nn.relu),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Conv2D(64, (3,3), activation=tf.nn.relu),
    tf.keras.layers.MaxPooling2D(2, 2),
    
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(512, activation=tf.nn.relu),
    tf.keras.layers.Dense(1, activation=tf.nn.sigmoid)
])

model_2.summary()
```

## Train model_1

```python
%%time

model_1.compile(
    optimizer=tf.keras.optimizers.RMSprop(learning_rate=0.001),
    loss=tf.keras.losses.binary_crossentropy,
    metrics=["accuracy"]
)

history_model1 = model_1.fit(
    train_gen_150px,
    steps_per_epoch=8,
    epochs=10,
    validation_data=val_gen_150px,
    validation_steps=8,
    verbose=1
)
```

```python
model_2.compile(
    optimizer=tf.keras.optimizers.RMSprop(learning_rate=0.001),
    loss=tf.keras.losses.binary_crossentropy,
    metrics=["accuracy"]
)

history_model2 = model_2.fit(
    train_gen_150px,
    steps_per_epoch=8,
    epochs=15,
    validation_data=val_gen_150px,
    validation_steps=8,
    verbose=1
)
```

```python
plt.plot(history_model1.history["loss"])
plt.plot(history_model1.history["val_loss"])
plt.plot(history_model1.history["accuracy"])
plt.plot(history_model1.history["val_accuracy"])
plt.legend(["loss","val_loss","accuracy","val_accuracy"])
plt.show();
```

```python
plt.plot(history_model2.history["loss"])
plt.plot(history_model2.history["val_loss"])
plt.plot(history_model2.history["accuracy"])
plt.plot(history_model2.history["val_accuracy"])
plt.legend(["loss","val_loss","accuracy","val_accuracy"])
plt.show();
```

### Model 3 - 5 convolutional layers and 300x300px input shape

```python
import tensorflow as tf

model = tf.keras.models.Sequential([
    # Note the input shape is the desired size of the image 300x300 with 3 bytes color
    # This is the first convolution
    tf.keras.layers.Conv2D(16, (3,3), activation='relu', input_shape=(300, 300, 3)),
    tf.keras.layers.MaxPooling2D(2, 2),
    # The second convolution
    tf.keras.layers.Conv2D(32, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    # The third convolution
    tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    # The fourth convolution
    tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    # The fifth convolution
    tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    
    # Flatten the results to feed into a DNN
    tf.keras.layers.Flatten(),
    # 512 neuron hidden layer
    tf.keras.layers.Dense(512, activation='relu'),
    # Only 1 output neuron. It will contain a value from 0-1 where 0 for 1 class ('horses') and 1 for the other ('humans')
    tf.keras.layers.Dense(1, activation='sigmoid')
])
```
