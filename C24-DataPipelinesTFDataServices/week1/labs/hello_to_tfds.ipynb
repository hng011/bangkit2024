# ETL (Extract -> Transform -> Load)

```python
# Extract 
ds = tfds.load(name="mnist", split="train")

# Transform
ds.shuffle(100)

# Load
for d in ds.take(10):
    img = d["image"].numpy().squeeze()
    lbl = d["label"].numpy()
    
    print(f"Label: {lbl}")
    plt.imshow(img, cmap=plt.cm.binary)
    plt.show()
```
